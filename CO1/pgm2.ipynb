```python
import matplotlib.pyplot as plt
x=[2,6,4,8,5]
y=[3,1,6,4,2]
plt.plot(x, label='time', color='green')
plt.xlabel('x-axis')
plt.title('Time-Speed Graph')
plt.plot(y, label='speed')
plt.ylabel('y-axis')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([2,9,4,1])
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, 'o:b')
plt.plot(xpoints, 'o:g')
plt.show()
```
